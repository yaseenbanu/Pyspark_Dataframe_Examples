Task: Write a Spark/SQL code to make a Pair.

```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName('challenges').getOrCreate()
```

```python
# Define the input data
data = [("INDIA",),
        ("CHINA",),
        ("PAKISTAN",)]

# Create a DataFrame
df = spark.createDataFrame(data, ["Country"])

df.show()
```

```python
from pyspark.sql.functions import *

# Alias the DataFrame for joining and select the 'Country' column with appropriate aliases
df1 = df.alias('df1').select(col('Country').alias('PAIR1'))
df2 = df.alias('df2').select(col('Country').alias('PAIR2'))

# Perform inner join and filter out rows where countries are equal
joined_df = df1.join(df2, on= df1.PAIR1 < df2.PAIR2, how='inner')

# Show the joined DataFrame
joined_df.show()
```
